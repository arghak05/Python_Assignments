3. Describe three different data types.

Complex Numbers : Complex number is represented by complex class. It is specified as (real part) + (imaginary part)j. 
Eg : 2+3j

Tuple : Tuple is an ordered collection of Python objects. Tuples are always represented in parenthesis(). Tuples are immutable i.e. tuples cannot be modified after it is created. It is represented by tuple class.
Eg : t = (1,2,3,4)

Dictionary : Dictionary is an unordered collection of data values, used to store data values like a map. Dictionary holds key-value pair. Tuples are always represented in curly braces{}.
Eg : d = {"k1":"key1", "k2":"key2"}
Here, "k1" and "k2" are the keys and "key1" and "key2" are their values respectively.

```python
print("Complex Number:",type(2+3j))
print("Tuple:",type((1,2,3,4)))
print("Dictionary:",type({"k1":"key1", "k2":"key2"}))
```
