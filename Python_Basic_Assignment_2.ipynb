10. How can you tell the difference between break and continue?

The break statement stops the loop in which the statement is placed. When a break statement is executed, the statements after the contents of the loop are executed. 
The continue statement instructs a loop to continue to the next iteration. Any code that follows the continue statement is not executed. Unlike a break statement, a continue statement does not completely halt a loop.

```python
#break statement
for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
#continue statement
for i in range(10):
    if i == 5:
        continue
    print(i)
```
